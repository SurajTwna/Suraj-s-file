```python
import matplotlib.pyplot as plt
import numpy as np

graph = input("choose from: Sine, Cos, Tan")

def sine_graph():
  x = np.linspace(-3*np.pi, 3*np.pi, 100) #This allows the numbers on the x axis to start at - 2 and end at 2 and the 1000 allows for the points to be equally separated to enhance visuals
  y = np.sin(x) # this line plugs in x into the y value with the function being np.sin becasue it is a sin graph, and the same as the rest
  plt.grid(True)
  plt.title('Sine Graph For Population Data') #the last lines are basically the grid presentation and how I plot the graph
  plt.plot(x, y)
  plt.show()


def tan_graph():
  x = np.linspace(-3*np.pi, 3*np.pi, 100)
  y = np.tan(x)
  plt.grid(True)
  plt.title('Tan Graph For Population Data')
  plt.plot(x, y)
  plt.ylim(-4,4) #this y lim function sets the y values on the grid to only be showed at 4 so you can clearly see the tan graph
  plt.show


def cos_graph():
  x = np.linspace(-3*np.pi, 3*np.pi, 100)
  y = np.cos(x)
  plt.grid(True)
  plt.title('Cos Graph For Population Data')
  plt.plot(x, y)
  plt.show()


if graph == "sine": #this code block sets conditions so that if the names of the graphs are inputed it prints the graph
  sine_graph()

elif graph == "tan":
  tan_graph()

elif graph == "cos" :
  cos_graph()
else:
  print("Redo your choice from tan, sine, and cos")
```
